## Random Forest Regression
### Import libraries & dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/Users/liakooras/Desktop/UDEMIES/MACHINE LEARNING A-Z/Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 9 - Random Forest Regression/Python/Position_Salaries.csv')
dataset.head()
```

```python
X = dataset.iloc[:,1:2].values
y = dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

### Training the RFR model into the whole dataset

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators= 10, random_state = 0)
```

```python
regressor.fit(X,y)
```

### Predicting a new result

```python
regressor.predict([[6.5]])
```

### Visualising the RFR results (higher resolution)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X,y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Random Forest Regression model prediction')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
